```python
import pandas as pd
```

```python
df = pd.read_csv("Walmart.csv")
```

```python
df.head()
```

```python
# Check the size of the dataset: number of rows and columns
df.shape
```

```python
# Get summary statistics for numerical columns to understand data distribution
df.describe()
```

```python
# Counts the duplicate values in dataset
df.duplicated().sum()
```

```python
# Remove duplicates from dataset
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df.shape
```

```python
# Counts the duplicate values in dataset
df.isnull().sum()
```

```python
#droppping all rows with missing records
df.dropna(inplace=True)

# verify
df.isnull().sum()
```

```python
df.shape
```

```python
# Check the data type of each column to understand the structure of the dataset
df.dtypes
```

```python
#Converts data type for unit price
df['unit_price'] = df['unit_price'].str.replace('$', '').astype(float)

df.head()
```

```python
# Get a concise summary of the dataset, including column types and non-null counts
df.info()
```

```python
df.columns
```

```python
df['total'] = df['unit_price'] * df['quantity']
df.head()
```

```python
# Get all column names in the dataset
df.columns
```

```python
#Creates a new column total_price
df['total_price'] = df['unit_price'] * df['quantity']
df.head()
```

```python
#Fixing column names to lower case
df.columns = df.columns.str.lower()
df.columns
```

```python
df.shape
```

```python
#Save dataset
df.to_csv('walmart_clean_data.csv', index=False)
```

# Connecting python script to mysql

```python
from sqlalchemy import create_engine

# MySQL connection
username = "root"
password = "pass123"
host = "localhost"
port = "3306"
database = "walmart_sales_db"

engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}")

# Write DataFrame to MySQL
table_name = "walmart_sales"   
df.to_sql(table_name, engine, if_exists="replace", index=False)

# Read back sample
pd.read_sql("SELECT * FROM walmart_sales LIMIT 5;", engine)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")
```
